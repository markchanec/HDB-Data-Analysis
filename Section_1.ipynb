```python
import pandas as pd
#load
df = pd.read_csv("data/coords/coords_combine.csv")
```

```python
# df = df.reset_index()
df.drop(['storey_range', 'flat_model', 'mth_sold', 'age_sold'], axis=1, inplace=True)
#merge back masionette into executive flat_type
df.loc[df.flat_type == 'MAISONETTE', 'flat_type'] = 'EXECUTIVE'
df.head()
```

```python
#extract only relevant info for question
temp = df.filter(['town','flat_type', 'floor_area_sqm', 'resale_price', 'year_sold'], axis=1)
s1question_B = temp.groupby(by=['town', 'flat_type', 'year_sold'], as_index=False).median()
```

```python
s1question_B
```

```python
#save file
s1question_B.to_csv("data/s1question_B.csv", index=False)
```
